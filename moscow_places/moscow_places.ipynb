## Исследовательский анализ

### Категории заведений

```python
# зададим формулу подписи на графиках
def show_values(axs, orient="v", space=.03):
    def _single(ax):
        if orient == "v":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() / 2
                _y = p.get_y() + p.get_height() + (p.get_height()*0.01)
                value = '{:.2f}'.format(p.get_height())
                ax.text(_x, _y, value, ha="center") 
        elif orient == "h":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() + float(space)
                _y = p.get_y() + p.get_height() - (p.get_height()*0.5)
                value = '{:.0f}'.format(p.get_width())
                ax.text(_x, _y, value, ha="left")

    if isinstance(axs, np.ndarray):
        for idx, ax in np.ndenumerate(axs):
            _single(ax)
    else:
        _single(axs)
```

```python
# создадим таблицу с количеством объектов и медианным значением посадочных мест по категориям заведений
df_category = (df.groupby('category').agg({'category':'count', 'seats':'median'})
      .rename(columns={'category': 'count', 'seats':'seats_median'}).reset_index())
```

```python
# зададим формулу для построения графиков по категориям
def ratio_chart(data1, x, y, palette, orient):
    sns.set_style('white')
    plt.figure(figsize=(8, 5))
    ax = sns.barplot(x= x, y= y, data=data1, palette = palette) 
    show_values(ax, orient)
    plt.grid()
```

```python
ratio_chart(df_category, 'count', 'category','Spectral','h')
plt.title('Количество объектов общественного питания по категориям')
plt.xlabel('Количество заведений')
plt.ylabel('Категория заведения')
plt.show()
```

В данных представлено 8 категорий. Наибольшее количество заведений представляют собой кафе, на втором месте по количеству заведений идет категория ресторанов, меньше всего столовых и булочных.

### Топ 15 улиц по количеству заведений

```python
# создадим таблицу с названием топ 15 улиц
df_top_streets = (df.groupby('street').agg({'name':'count'})
               .rename(columns = ({'name':'count'}))
               .sort_values(by='count', ascending = False)
               .reset_index()
               .head(15))
# сохраним в переменной названия улиц
query_street = df_top_streets['street']
# получим из основной таблицы таблицу только с топ 15 популярными улицами
pop_street = df.query('street in @query_street')
```

```python
# теперь создадим сводную таблицу категорий заведений для каждой улицы из топа и построим по ней график
popylar_street_pivot = (pop_street.pivot_table(index ='street', columns = 'category',values = 'address',
                                               aggfunc = 'count').fillna(0)
                        .plot.bar(stacked=True, figsize=(16,10)))
popylar_street_pivot.set_title('График распределения количества заведений и их категорий по "топ 15" улицам')
popylar_street_pivot.set_xlabel ('Улица')
popylar_street_pivot.set_ylabel('Количество заведений, шт.')
popylar_street_pivot.legend(loc=2, title = 'Категория:')
plt.ylim(0, 200)
plt.grid()
plt.show()
```

Больше всего заведений располанается на проспекте Мира, там находится порядка 180 заведений, следующей с существенным отрывом(122 заведения) на Профсоюзной улице. Обе эти улицы имеют длину около 9 км, неудивительно, что на них расположено так много заведений. Да и в принципе в подборке топ 15 улиц практически все они достаточно протяженные, за исключением Пятницкой улицы, находящейся в цетнтре города и являющейся довольно богатой на различного типа заведения.

Так же на подовляющем большинстве улиц по количеству заведений лидируют категории кафе, кофейни или рестораны.
