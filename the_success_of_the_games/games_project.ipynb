### Изучение типов данных в столбцах и работа с пропусками

Рассмотрим типы данных в каждом столбце:
- Найдём столбцы, в которых нужно изменить тип данных.
- Преобразуем тип данных в выбранных столбцах.

````python
print(data.info()) # вспомним какие типы об'ектов хранит датасет
````

В столбце `name` содержится два пропуска, выведем их на экран.

````python
display(data[data['name'].isna() == True])
````

Заменить данные по названию игры ничем невозможно, придется такие строки удалить, тем более, что их всего 2.

````python
# посмотрим какие значения прописаны в столбце 'year_of_realese'
print(data['year_of_release'].sort_values().unique())
````

В столбце `year_of_release` содержится только год выпуска игры с 1980 по 2016 включительно.\
ропусков в столбце 269, что составляет 1,6% от всего датасета. Данные по году выпуска игры в количестве 269 штук восстановить проблематично, поэтому от строк, где не указан год выпуска придется избавиться.
Столбец `year_of_release` имеет тип данных 'float64'. Менять тип данных на 'dstetime' не имеет смысла, т.к. указан только год выпуска игр, однако тип данных можно поменять на 'int' для дальнейшего удобства восприятия, ведь год выпуска игр может быть представлен исключительно целым числом.

````python
# удалим строки, для которых пропущены значения в столбцах 'name' и 'year_of_release'
data=data.dropna(subset=['name','year_of_release']).reset_index(drop=True) 
# заменим тип данных в строке year_of_release' на 'int'
data['year_of_release']=data['year_of_release'].astype('int32')
````

Посмотрим, какие значения представлены в столбце `rating`

````python
# выведем на экран уникальные значения столбца 'rating' и их количество
display(data['rating'].value_counts())
````

````python
# посмотрим, что за единственная игра с только ожидающимся рейтингом
display(data[data['rating']=='RP'])
````

По данным всемирной паутины: ESRB была основана в 1994 году ассоциацией Interactive Digital Software Association (современное название — «Entertainment Software Association»). К началу 2003 года рейтинги ESRB получили более 8000 продуктов от 350 издателей.

Рейтинг E первоначально назывался K-A. Заменим значение 'K-A' на 'E' в датасете, чтобы избежать путаницы.

Мы уже знаем, что в датасете представлены игры с 1980 по 2016 гг. Основанная в 1994 году компания могла попросту не присвоить некоторым играм рейтинги по разным причинам. Так же пропуски могли возникнуть в связи с тем, что ESRB — негосударственная организация, которая  принимает и определяет рейтинги для компьютерных видеоигр и другого развлекательного программного обеспечения в **США и Канаде**. Получается, что для игр, выпущенных в других странах, этот рейтинг не присваивается.
Удалять пропуски не будем, заменить их ничем вразумительным не представляется возможным, присвоим пропускам значение "unknown".

Остается две группы значений по одной игре в каждой.\
Это «RP» («Rating Pending») — «Рейтинг ожидается» и «AO» («Adults Only 18+») — «Только для взрослых». Поместим "AO" в "M"(для взрослых). 
Согласно сайту www.esrb.org на сегодняшний день игре Supreme Ruler: Cold War присвоен рейтинг "E10+".\
Выборка по играм с рейтингом "EC" безусловно не является репрезентативной, однако сделать с этим ничего нельзя.

````python
# заменим в рейтинге значение 'K-A' на 'E'
data['rating'] = data['rating'].str.replace('K-A', "E")
data['rating'] = data['rating'].str.replace('AO', "M")
data['rating'] = data['rating'].str.replace('RP', "E10+")
# заменим пропуски на значение 'unknown'
data['rating'] = data['rating'].fillna('unknown')
# проверим результат
print(data['rating'].unique())
````

Посмотрим, какие значения таит в себе столбец `user_score`

````python
# выведем на экран уникальные значения столбца 'user_score'
print(data['user_score'].sort_values().unique())
````

При знакомстве с датасетом, стало известно, что тип столбца `user_score`  - object. Не удивительно, ведь вместе с привычными нам численными значениями и пропусками здесь встречается значение 'tbd'. Данная аббревиатура расшифровывается как  - To Be Determined, то есть "Будет определено".

Видимо, сбор информации по некоторым играм проходил в период их анонса и информации о оценке пользователей еще не поступило. 

К сожалению, такие данные нельзя ничем заменить. В данной работе преобразуем их в пропуски, ведь никакой информации об оценке игры пользователями аббревиатура tbd не несет.

Что касается многочисленных пропусков в столбце, то данные пропуски могут быть связаны скорее с техническими причинами. 

````python
data.loc[data['user_score'] == 'tbd', 'user_score'] = np.nan
data['user_score'] = data['user_score'].astype('float')
print(data['user_score'].sort_values().unique())
````

Пропуски в столбцах `user_score` и `critic_score` нечем заменить. Здесь не подойдет заполнение пропусков ни медианным, ни средним, ни каким-либо другим значением.

### Анализ платформ с наибольшими продажами

````python
#построим график распределения платформ по суммарным продажам
game_total_sale = data.pivot_table(index ='platform', values = 'total_sales', aggfunc = 'sum').reset_index().\
                  sort_values(by='total_sales', ascending = False).\
                  plot.bar(x = 'platform', y = 'total_sales', figsize=(15,6), color = '#1A4780', grid=True)
game_total_sale.set_title('Распределение платформ по суммарным продажам')
game_total_sale.set_xlabel ('платформа')
game_total_sale.set_ylabel('суммарные прдажи, млн.копий')
plt.show()
````

Из графика распределения платформ по сумарным продажам явно выделяются 6 лидеров по продажам игр: DS, PS, PS2, PS3, Wii, X360.\
Выбрав из общего датасета только вышеуказанные платформы, посмотри, как распределялись продажи игр на эти платформы по годам.

````python
# для начала создадим таблицу, в которой будут только 6 "платформ-лидеров"
game_platform_year= data.loc[data['platform'].isin(['DS','PS','PS2','PS3','Wii','X360'])]
# теперь создадим сводную таблицу продаж игр на каждую платформу за каждый год и построим по ней график
gp_year_table = game_platform_year.pivot_table(index ='year_of_release', columns = 'platform',\
                                     values = 'total_sales', aggfunc = 'sum').\
                                   plot.bar(stacked=True, figsize=(15,6))
gp_year_table.set_title('Распределение продаж игр на платформы по годам')
gp_year_table.set_xlabel ('год')
gp_year_table.set_ylabel('суммарные прдажи, млн.копий')
gp_year_table.legend(loc=2, title = 'Платформа')
plt.show()
````

По графику выше видо, для каждой платформы существует 10-12-летний период ее рентабельности. Затем игры на платформу перестают выпускать и на смену приходят новинки производства.\
Получается, для того, чтобы сделать прогноз на 2017 год, достаточным будет проанализировать данные за последние 5 лет. В срезе таких данных мы сможем сориентироваться в какой фазе "жизненного пути" находится та или иная платформа.

###  Анализ платформ за актуальный период времени

````python
#создадим датасет с играми старше 20011 года
data_current = data.loc[data['year_of_release'] > 2011]
````

Посмотрим, какие платформы за актуальный период лидируют по продажам.

````python
# создадим сводную таблицу продаж игр на каждую платформу с 2012 по 2016 гг и построим по ней график
dc_table = data_current.pivot_table(index ='year_of_release', columns = 'platform',\
                                     values = 'total_sales', aggfunc = 'sum').\
                                   plot.bar(stacked=True, figsize=(15,4))
dc_table.set_title('Распределение продаж игр на платформы по годам')
dc_table.set_xlabel ('год')
dc_table.set_ylabel('суммарные прдажи, млн.копий')
dc_table.legend (title = 'Платформа')
plt.show()
````

Помня о том, что данные на 2016 год могут быть не полными, из вышепредставленного графика можно сделать следующие выводы:
 - PS4, выпущенная в 2013 году только набирает популярность, заменяя в предыдущие годы свою предшественницу PS3
 - портативная консоль 3DS к концу исследуемого периода испытывает уже скорее падение продаж по играм
 - XOne, выпущенный в 2013 году так же как и PS4 к 2017 году должен находится в пике продаж по играм 
 - Cуммарные продажи на ПК падют, несмотря на разнообразие игровых компьютеров
 - Смотря на столбец по 2015 году можно выделить несколько лидеров среди платформ: PS4 и XOne. Так же в потенциально прибыльные можно определить 3DS.

## Проверка гипотез

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

При проверке обеих гипотез будем использовался метод для проверки гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них выборкам.\
Будем применять данный метод т.к. у нас нет информации об оценках всех пользователей и в нашем распоряжении только некоторые выборки.\
Для того, чтобы метод проверки гипотиз работал корректно, создадим две отдельные таблицы по интересующим нас параметрам, а затем удалим из них строки с отсутствующим рейтингом.\
Для каждой гипотезы проверим размеры выборок, если размеры получатся разными, тогда укажем в тесте equal_var = False.

```
H_0:
Средний польовательский рейтинг (user_score) платформы XOne = cреднему польовательскому рейтингу (user_score) платформы PC`

H_a:
Средний польовательский рейтинг (user_score) платформы XOne ≠ cреднему польовательскому рейтингу (user_score) платформы PC`

alpha = 0.05
```

````python
# создадим таблицы с интересующими нас выборками
xone = data_current.loc[data_current['platform'] =='XOne'].dropna(subset=['user_score'])
pc = data_current.loc[data_current['platform'] =='PC'].dropna(subset=['user_score'])
````

````python
# проверим размеры выборок
print('Размер таблицы xone:', xone.shape)
print('Размер таблицы pc:', pc.shape)
````

````python
# выведем на экран средние значения наших выборок
print('Среднее значение пользовательского рейтинга XOne:', xone['user_score'].median())
print('Среднее значение пользовательского рейтинга PC:',pc['user_score'].median())
````

Размеры выборок разные, укажем в тесте `equal_var = False`.

````python
#вызов метода для проверки гипотезы
results = st.ttest_ind(xone['user_score'],pc['user_score'], equal_var = False)

#зададим значение уровня значимости   
alpha = 0.05 

# выведем значения p-value на экран
print(results.pvalue)  

# Выведем на экран результат теста
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else: print('Не получилось отвергнуть нулевую гипотезу')
````

В ходе проверки первой гипотезы не получилось опровергнуть нулевую гипотезу о том, что средние пользовательские рейтинги платформ Xbox One и PC одинаковые, т.к. значение p-value достаточно высоко.
Что подтверждается средними пользовательскими рейтингами по платформам

### Средние пользовательские рейтинги жанров Action и Sports  разные.

Мы предполагаем, что средние пользовательские рейтинги жанров Action и Sports разные.
Однако при составлении гипотез нам следует сформулировать нулевую гипотезу как средние пользовательские рейтинги жанров Action и Sports одинаковые.

```
H_0:
Средний польовательский рейтинг (user_score) жанра Action = cреднему польовательскому рейтингу (user_score) жанра Sports`

H_a:
Средний польовательский рейтинг (user_score) жанра Action ≠ cреднему польовательскому рейтингу (user_score) жанра Sports`

alpha = 0.05
```

````python
# создадим таблицы с интересующими нас выборками
action = data_current.loc[data_current['genre'] =='Action'].dropna(subset=['user_score'])
sports = data_current.loc[data_current['genre'] =='Sports'].dropna(subset=['user_score'])
````

````python
# проверим размеры выборок
print('Размер таблицы action:', action.shape)
print('Размер таблицы sports:', sports.shape)
````

````python
# выведем на экран средние значения наших выборок
print('Среднее значение пользовательского рейтинга жанра "Action":', action['user_score'].median())
print('Среднее значение пользовательского рейтинга жанра "Sports":',sports['user_score'].median())
````

Размеры выборок разные, укажем в тесте `equal_var = False`.

````python
#вызовем метода для проверки гипотезы
results = st.ttest_ind(action['user_score'], sports['user_score'], equal_var = False)

#зададим значение уровня значимости 
alpha = 0.05 

# выведем значения p-value на экран
print(results.pvalue)  

# Выведем на экран результат теста
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else: print('Не получилось отвергнуть нулевую гипотезу')
````

После проведения теста отвергнута нулевая гипотезу о том, что средние пользовательские рейтинги жанров Action и Sports одинаковы, т.к. значение p-value достаточно низкое.
Что подтверждается средними пользовательскими рейтингами по данным жанрам.

Для того, чтобы сформулировать нулевую гипотезу в нашем случае необходимо понять среднее каких двух ГС мы сравниваем. После этого формулируем нулевую гипотезу как: средние двух генеральных совокупностей равны между собой. Альтернативная гипотеза формулируется из отрицания нулевой - средние не равны.

В первом случае не получилось отвергуть нулевую гипотезу, значит мы остались при своём предположении, что средние генеральных совокупностей равны.

Во втором случае мы предполагаем, что средние двух ГС не равны, при проведении теста мы отвергли нулевую гипотезу о том, что средние двух ГС равны, а, значит, снова остались при своём изначальном предположении, что средние этих ГС не равны.

Для проверки гипотез применялся t-критерий Стьюдента для независимых выборок. Критерий был выбран т.к. он применяется для проверки гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них выборкам.
В Python  такую задачу решает метод  scipy.stats.ttest_ind().
