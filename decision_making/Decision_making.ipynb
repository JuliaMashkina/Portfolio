## Приоритизация гипотез.

### Фреймворк ICE 

Рассчитаем ICE для приоритизации гипотез

***ICE SCORE*** = `(Impact*Confidence)/Efforts`

````python
pd.options.display.max_colwidth = 1
hypothesis['ice'] = round((hypothesis['impact'] * hypothesis['confidence'])
                          /hypothesis['efforts'])

display(hypothesis[['hypothesis', 'ice']]
        .sort_values(by='ice', ascending = False))
````

Наиболее перспективными являются гипотезы:\
8. Запустить акцию, дающую скидку на товар в день рождения;\
0. Добавить два новых канала привлечения трафика, что позволит привлекать на 30% больше пользователей;\
7. Добавить форму подписки на все основные страницы, чтобы собрать базу клиентов для email-рассылок.

###  Фреймворк RICE 

Рассчитаем RICE для приоритизации гипотез

***RICE SCORE*** = `(Reach*Impact*Confidence)/Efforts`

````python
hypothesis['rice'] = round((hypothesis['reach'] * hypothesis['impact'] * hypothesis['confidence'])
                           /hypothesis['efforts'])

display(hypothesis[['hypothesis', 'reach','rice']]
        .sort_values(by='rice', ascending = False))
````

При рассчете фреймворка RICE в отличии от ICE учитывается еще и охват пользователей.

Так, наиболее перспекттивны гипотезы:\
7. Добавить форму подписки на все основные страницы, чтобы собрать базу клиентов для email-рассылок;\
2. Добавить блоки рекомендаций товаров на сайт интернет магазина, чтобы повысить конверсию и средний чек заказа;\
0. Добавить два новых канала привлечения трафика, что позволит привлекать на 30% больше пользователей;\
6. Показать на главной странице баннеры с актуальными акциями и распродажами, чтобы увеличить конверсию.

**Таким образом,** при рассчете двух фреймворков приоритезация гипотез сменилась. Это произошло ввиду высокого значения Reach у гипотезы 7 и минимального значения этого параметра у гипотезы 8, которая изначально была наиболее перспективной. 
Так же достаточно высокое значение Reach у гипотезы 2, поэтому она так же сильно поднялась в рейтинге.


## Анализ A/B-теста

Чтобы построить куммулятивные графики, для начала нужно собрать данные с накоплением по каждой дате для каждой группы.

````python
# создаем массив уникальных пар значений дат и групп теста
dates_groups = orders[['date','group']].drop_duplicates() 

# получаем агрегированные кумулятивные по дням данные о заказах 
orders_aggregated = (dates_groups
                    .apply(lambda x: orders[np.logical_and(orders['date']<= x['date'],
                                                           orders['group'] == x['group'])]
                           .agg({'date' : 'max',
                                 'group' : 'max',
                                 'transactionid' : 'nunique',
                                 'visitorid' : 'nunique',
                                 'revenue' : 'sum'}),axis=1)
                    .sort_values(by=['date','group']))

# получаем агрегированные кумулятивные по дням данные о посетителях
visitors_aggregated = (dates_groups
                      .apply(lambda x: visitors[np.logical_and(visitors['date'] <= x['date'],
                                                               visitors['group'] == x['group'])]
                             .agg({'date' : 'max',
                                   'group' : 'max',
                                   'visitors' : 'sum'}), axis=1)
                      .sort_values(by=['date','group']))

# объединяем кумулятивные данные в одной таблице и присваиваем ее столбцам понятные названия
cumulative_data = orders_aggregated.merge(visitors_aggregated, left_on=['date', 'group'], right_on=['date', 'group'])

cumulative_data.columns = ['date', 'group', 'orders', 'buyers', 'revenue', 'visitors']

display(cumulative_data.head(5)) 
````

### Куммулятивные метрики

#### График кумулятивной выручки по группам

````python
# датафрейм с кумулятивным количеством заказов и кумулятивной выручкой по дням в группе А
cumulative_revenue_a = cumulative_data[cumulative_data['group']=='A'][['date','revenue', 'orders']]

# датафрейм с кумулятивным количеством заказов и кумулятивной выручкой по дням в группе B
cumulative_revenue_b = cumulative_data[cumulative_data['group']=='B'][['date','revenue', 'orders']]

plt.figure(figsize=[8, 4])

# Строим график выручки группы А
plt.plot(cumulative_revenue_a['date'], cumulative_revenue_a['revenue'], label='A',)

# Строим график выручки группы B
plt.plot(cumulative_revenue_b['date'], cumulative_revenue_b['revenue'], label='B')

plt.xticks(rotation=20)
plt.title('Куммулятивная выручка по группам')
plt.xlabel ('дата')
plt.ylabel('суммарная выручка, млн.')
plt.legend() 
plt.show()
````

Выручка увеличивается на протяжении всего теста. График группы В резко растет в районе 17 августа 2008. Вероятно, в это время произошел всплеск заказон в этой группе. График группы А равномерный на протяжении всего времени. 

#### График кумулятивного среднего чека по группам

````python
plt.figure(figsize=[8, 4])
# Строим график среднего чека группы А
plt.plot(cumulative_revenue_a['date'],
         cumulative_revenue_a['revenue']/cumulative_revenue_a['orders'],
         label='A')
# Строим график среднего чека группы В
plt.plot(cumulative_revenue_b['date'],
         cumulative_revenue_b['revenue']/cumulative_revenue_b['orders'],
         label='B')

#задаем параметры визуализации
plt.xticks(rotation=20)
plt.title('Куммулятивный средний чек по группам')
plt.xlabel ('дата')
plt.ylabel('средний чек')
plt.legend() 
plt.show()
````

Средний чек группы А к концу теста становится равномерным и устанавливается в районе 7000.\
Средний чек группы В неравномерен. В районе 17-го августа наблюдается резкий скачок среднейго чека с ~7000 до ~ 10000. Далее значения среднего чека имебт отрицательный тренд и к концу теста достигают уже ~9000.

#### Относительное изменения кумулятивного среднего чека группы B к группе A

````python
# задаем размер графика
plt.figure(figsize=[8, 4])

# собираем данные в одном датафрейме
merged_cumulative_revenue = cumulative_revenue_a.merge(cumulative_revenue_b,
                                                   left_on='date',
                                                   right_on='date',
                                                   how='left',
                                                   suffixes=['A', 'B'])

# cтроим отношение средних чеков
plt.plot(merged_cumulative_revenue['date'],
         (merged_cumulative_revenue['revenueB']/merged_cumulative_revenue['ordersB'])
          /(merged_cumulative_revenue['revenueA']/merged_cumulative_revenue['ordersA'])
         -1, color = '#B32821')

# добавляем ось X
plt.axhline(y=0, color='black', linestyle='--')
#задаем параметры визуализации
plt.xticks(rotation=20)
plt.title('Изменение кумулятивного среднего чека группы B к группе A')
plt.xlabel ('дата')
plt.ylabel('доля')
plt.show()
````

График куммулятивного среднего чека группы В к группе А сильно скачет. Так в районе 13 августа показатели среднего чека группы В отлосительно А сильно ухудшились, а затем снова резко улучшились 17 августа, после чего стабильно ухудшаются.

#### График кумулятивного среднего количества заказов на посетителя по группам

````python
# считаем кумулятивную конверсию
cumulative_data['conversion'] = cumulative_data['orders']/cumulative_data['visitors']

# отделяем данные по группе A
cumulative_data_a = cumulative_data[cumulative_data['group']=='A']

# отделяем данные по группе B
cumulative_data_b = cumulative_data[cumulative_data['group']=='B']

# задаем размер графика
plt.figure(figsize=[8, 4])

# строим графики
plt.plot(cumulative_data_a['date'], cumulative_data_a['conversion'], label='A')
plt.plot(cumulative_data_b['date'], cumulative_data_b['conversion'], label='B')
plt.legend()

# зрассчитаем минимальную и максимальную дату исследования
min_data = cumulative_data['date'].min()
max_data = cumulative_data['date'].max()

# задаем масштаб осей
plt.axis([min_data, max_data, 0.02, 0.040])

plt.xticks(rotation=20)
plt.title('Изменение кумулятивного среднего количества заказов\n на посетителя по группам')
plt.xlabel ('дата')
plt.ylabel('доля')
plt.show()
````

До 5 августа лидировала группа A, после группа В вырвалась вперед и была стабильно выше А. Обе кривые стабилизируются к концу исследования.

#### Относительное изменения кумулятивного среднего количества заказов на посетителя группы B к группе A

````python
# считаем кумулятивную конверсию
mer_сumulative_сonv = (cumulative_data_a[['date','conversion']]
                               .merge(cumulative_data_b[['date','conversion']],
                                      left_on='date',
                                      right_on='date',
                                      how='left',
                                      suffixes=['A', 'B']))
# задаем размер графика
plt.figure(figsize=[8, 4])

#строим график 
plt.plot(mer_сumulative_сonv['date'],
         mer_сumulative_сonv['conversionB']/mer_сumulative_сonv['conversionA']-1,
         label="Относительный прирост конверсии\n группы B относительно группы A",
        color = '#B32821')
#задаем параметры визуализации
plt.legend()
plt.axhline(y=0, color='black', linestyle='--')
plt.axhline(y=0.15, color='grey', linestyle='--')
plt.axis([min_data, max_data, -0.4, 0.4])
plt.xticks(rotation=20)
plt.title('Изменение кумулятивного среднего количества заказов\n на посетителя группы B к группе A')
plt.xlabel ('дата')
plt.ylabel('доля')
plt.show()
````

Далее проведем А/В тестирование и выясним статистическую значимость различия между группами исследования по "сырым" данным, а так же по данным, лишенным выбросов.

Для исследования будем применять непараметрический тест Уилкоксона-Манна-Уитни. Выбор данного теста обусловлен наличием в данных больших выбросов. При применении параметрических тестов выполняется сравнение рангов одних и тех же значений, попавших в обе выборки.

### Статистическая значимость  по «сырым» данным

####  Различия в среднем количестве заказов на посетителя между группами

Сформулируем гипотезы:
```
H_0: Среднее количество заказов группы А = Среднее количество заказов группы В`
H_1: Среднее количество заказов группы А ≠ Среднее количество заказов группы В`
alpha = 0.05
```

````python
# создаем таблицу количества заказов пользователей группы А
orders_by_users_a = (
    orders[orders['group'] == 'A']
    .groupby('visitorid', as_index=False)
    .agg({'transactionid': pd.Series.nunique}))

orders_by_users_a.columns = ['userid', 'orders']

# создаем таблицу количества заказов пользователей группы В
orders_by_users_b = (
    orders[orders['group'] == 'B']
    .groupby('visitorid', as_index=False)
    .agg({'transactionid': pd.Series.nunique})
)
orders_by_users_b.columns = ['userid', 'orders']

"""" создадим семпл,
в котором к таблице количества заказов пользователей группы А
добавим пользователей не совершивших ни одной покупки """

sample_a= pd.concat([orders_by_users_a['orders'],
                     pd.Series(0,
                               index=np.arange(visitors.query('group == "A"')['visitors'].sum()
                                               - len(orders_by_users_a['orders'])),
                               name='orders')],
                    axis=0)
# проделаем предыдущую операцию для группы В
sample_b= pd.concat([orders_by_users_b['orders'],
                     pd.Series(0,
                               index=np.arange(visitors.query('group == "B"')['visitors'].sum()
                                               - len(orders_by_users_b['orders'])),
                               name='orders')],
                    axis=0)
# выведем на экран результаты теста Mann Whitneyu и относительный прирост среднего числа заказов группы B
print('p-value:',
      "{0:.3f}".format(stats.mannwhitneyu(sample_a, sample_b)[1]))

print('Относительный прирост среднего числа заказов группы B:',
      "{0:.3f}".format(sample_b.mean() / sample_a.mean() - 1))
````

p-value  меньше 0.05, значит отвергаем нулевую гипотезу о равенстве количества заказов по группам. Относительный выигрыш группы В составляет 13,8%.

#### Различия в среднем чеке заказа между группами

Сформулируем гипотезы:
```
H_0: Средний чек группы А = Средний чек группы В`
H_1: Средний чек группы А ≠ Средний чек группы В`
alpha = 0.05
```

````python
# выведем на экран Статистическая значимость и относительные различия в среднем чеке заказа между группами
print('p-value:',
    '{0:.3f}'
      .format(stats.mannwhitneyu(orders[orders['group']=='A']['revenue'],
                                 orders[orders['group']=='B']['revenue'])[1]))

print('Относительные различия в среднем чеке между группами:',
    '{0:.3f}'
      .format(orders[orders['group']=='B']['revenue'].mean()
              /orders[orders['group']=='A']['revenue'].mean()-1))
````

p-value  больше 0.05, значит не отвергаем нулевую гипотезу о равенстве среднего чека по группам.\
Средний чек группы В на 25,9% выше, чем в группе А.

### Статистическая значимость  по «очищенным» данным

Выясним суммарное количество аномальных пользователей 

````python
# создадим таблицу пользователей с аномальным количеством заказов
users_with_many_orders = pd.concat(
    [orders_by_users_a[orders_by_users_a['orders']
                       > np.percentile(orders_by_users['orders'], 99)]['userid'],
        orders_by_users_b[orders_by_users_b['orders']
                          > np.percentile(orders_by_users['orders'], 99)]['userid']],
    axis=0)
# создадим таблицу пользователей с аномальной суммой заказа
users_with_expensive_orders = orders[orders['revenue']
                                     > round(np.percentile(orders['revenue'], 99))]['visitorid']
# объединим созданные таблицы, удалив возможные дубликаты
abnormal_users = (pd.concat([users_with_many_orders, users_with_expensive_orders], axis=0)
                 .drop_duplicates()
                 .sort_values())

print('Аномальных пользователей:', abnormal_users.shape[0]) 
````

#### Различия в среднем количестве заказов на посетителя между группами 

Узнаем влияют ли выбросы на результат тестирования

Сформулируем гипотезы:
```
H_0: Среднее количество заказов группы А = Среднее количество заказов группы В`
H_1: Среднее количество заказов группы А ≠ Среднее количество заказов группы В`
alpha = 0.05
```

````python
# создадим отфильтрованные семплы для групп А и В
sample_a_filtered = pd.concat([orders_by_users_a[np.logical_not(orders_by_users_a['userid']
                                                             .isin(abnormal_users))]['orders'],
                               pd.Series(0,
                                         index=np.arange(visitors.query('group == "A"')['visitors'].sum()
                                                         - len(orders_by_users_a['orders'])),
                                         name='orders')],axis=0)

sample_b_filtered = pd.concat([orders_by_users_b[np.logical_not(orders_by_users_b['userid']
                                                             .isin(abnormal_users))]['orders'],
                               pd.Series(0,
                                         index=np.arange(visitors.query('group == "B"')['visitors'].sum()
                                                         - len(orders_by_users_b['orders'])),
                                         name='orders')],axis=0)
# выведем на экран результаты эксперимента
print('p-value:',
    '{0:.3f}'.format(stats.mannwhitneyu(sample_a_filtered, sample_b_filtered)[1]))
print('Относительный прирост среднего числа заказов группы B:',
    '{0:.3f}'.format(sample_b_filtered.mean()/sample_a_filtered.mean()-1))
````

Результаты по количеству заказов изменились не существенно

#### Различия в среднем чеке заказа между группами

Посмотрим повлияло ли исключение аномальных пользователей на средний чек

Сформулируем гипотезы:
```
H_0: Средний чек группы А = Средний чек группы В`
H_1: Средний чек группы А ≠ Средний чек группы В`
alpha = 0.05
```

````python
# выведем на экран результаты эксперимента
print('p-value:',
    '{0:.3f}'
      .format(stats.
              mannwhitneyu(orders[np.logical_and(orders['group'] == 'A',
                                                 np.logical_not(orders['visitorid']
                                                                .isin(abnormal_users)))]['revenue'],
                           orders[np.logical_and(orders['group'] == 'B',
                                                 np.logical_not(orders['visitorid']
                                                                .isin(abnormal_users)))]['revenue'])[1]))

print('Относительные различия в среднем чеке между группами:',
    "{0:.3f}".format(orders[np.logical_and(orders['group'] == 'B',
                                             np.logical_not(orders['visitorid']
                                                            .isin(abnormal_users)))]['revenue'].mean()
                       / orders[np.logical_and(orders['group'] == 'A',
                                               np.logical_not(orders['visitorid']
                                                              .isin(abnormal_users)))]['revenue'].mean()
                       - 1))
````

p-value увеличился относительно рассчитанного по сырым данным. Разница между группами сократилась с 25,9% до -6%.
